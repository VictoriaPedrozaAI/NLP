```python
import gradio as gr
```

```python
import random
```

```python
from transformers import AutoTokenizer
from transformers import pipeline
from transformers import AutoModelForSequenceClassification
```

```python
model_name = "yiyanghkust/finbert-tone"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

pipe = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
pipe("The company reported a strong increase in quarterly earnings.")
```

```python
sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

def financial_sentiment(text):
    result = sentiment_pipeline(text) [0]
    label = result ['label']
    score = result ['score']
    return label, score


desc = "This app **Indentify Financial Sentiment**, submit a *financial sentence* as input " \
       "and it will return the predicted sentiment as **positive**, **neutral**, or **negative**, " \
       "the model is based on a finBERT with financial data."

article = "<h3>How to Use:</h3>" \
          "<ul>" \
          "<li>Enter a sentence related to financial markets or news.</li>" \
          "<li>Click the 'Submit' button to see the results.</li>" \
          "<li>You will get the sentiment label first and its confidence score as the sencond result.</li>" \
          "</ul>"

demo =  gr.Interface(
    fn=financial_sentiment,
    inputs=gr.Textbox(label="Financial Text", placeholder="Enter financial news or comment here..."),
    outputs=gr.Textbox(label="Predicted Sentiment"),
    title="📈 Financial Sentiment Analyzer (FinBERT)",
    description=desc,
    article=article,
    theme=gr.themes.Soft(),
    allow_flagging="never"
)

demo.launch(share=True)
```
