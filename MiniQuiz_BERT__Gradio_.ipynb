```python
import gradio as gr
from transformers import pipeline

qa_pipeline = pipeline('question-answering')

pre_contexts = {
    "Natural Language Processing (NLP)": """Natural Language Processing (NLP) is a field of artificial intelligence that focuses on enabling computers to understand, interpret, and generate human language. It combines linguistics, computer science, and machine learning to process text and speech. Common applications include machine translation, sentiment analysis, chatbots, and text summarization.""",

    "Transformer Models": """Transformer models are a type of deep learning architecture introduced in 2017 by the paper 'Attention is All You Need'. Unlike previous models like RNNs, transformers use self-attention mechanisms to process sequences in parallel, making them highly effective for NLP tasks. BERT, GPT, and T5 are all based on the transformer architecture.""",

    "BERT Model": """BERT (Bidirectional Encoder Representations from Transformers) is a pre-trained NLP model developed by Google. It reads text bidirectionally and is fine-tuned for tasks like question answering and sentiment analysis. BERT revolutionized NLP by improving context understanding and has been widely adopted in many AI applications.""",

    "Tokenization": """Tokenization is a key preprocessing step in NLP where text is broken down into smaller units called tokens. These can be words, characters, or subwords. Tokenization helps models understand structure and meaning in language. For example, the sentence 'NLP is fun!' can be tokenized into ['NLP', 'is', 'fun', '!'].""",
}

context_options = list(pre_contexts.keys())

def answer_question(context, question):
    result = qa_pipeline(question=question, context=context)
    return result['answer']

def update_context(choice):
    return pre_contexts.get(choice, "")

with gr.Blocks(title="Mini BERT Quiz App") as demo:
    gr.Markdown("# Mini BERT Quiz App")
    gr.Markdown("Ask BERT a question based on a preset paragraph or add a new one!")

    # Dropdown¿
    with gr.Row():
        context_dropdown = gr.Dropdown(
            choices=context_options,
            label="Select a Preset Text about NLP topics",
            info="By choosing an option the text will be filled automatically."
        )

    # Textbox for Context
    context_textbox = gr.Textbox(
        lines=8,
        label="Context (Paragraph)",
        placeholder="Enter or select a context for the BERT model to analyze"
    )

    #  Questions
    question_textbox = gr.Textbox(
        lines=2,
        label="Your Question",
        placeholder="Enter your questionto BERT model here."
    )

    # Answer
    output_answer = gr.Textbox(
        label="Answer",
        interactive=False,
        lines=2
    )


    submit_button = gr.Button("Get Answer")


    context_dropdown.change(
        fn=update_context,
        inputs=context_dropdown,
        outputs=context_textbox
    )

    submit_button.click(
        fn=answer_question,
        inputs=[context_textbox, question_textbox],
        outputs=output_answer
    )

demo.launch()
```
